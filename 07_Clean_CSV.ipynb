```python
import os
import pandas as pd

def replace_value_in_csv(folder_path, old_value, new_value):
    
    for filename in os.listdir(folder_path):
        if filename.endswith(".csv"):
            file_path = os.path.join(folder_path, filename)
            print (file_path)

            # Read the CSV file into a DataFrame
            df = pd.read_csv(file_path)

            # Replace the specified value in the entire DataFrame
            df.replace(old_value, new_value, regex=True, inplace=True)

            # Save the modified DataFrame back to the CSV file
            df.to_csv(file_path, index=False)

            print(f"Replaced '{old_value}' with '{new_value}' in {filename}")


# Example usage
folder_path = 'C:/Users/barth/Documents/LDACA/AusESL/clean_csv/'
old_value = 'LaBB-CAT_PNT:'
new_value = 'PNT:'

replace_value_in_csv(folder_path, old_value, new_value)
```
